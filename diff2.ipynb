```python
import sympy as sp
import math
from IPython.display import display, Math

x, t, xi, tau = sp.symbols('x t xi tau')

#Define functions and parameter a
a = sp.sqrt(10)
phi = x
psi = 0
f = 6*x**2*t
#

I1 = (1/2) * (phi.subs(x, x + a * t) + phi.subs(x, x - a * t))
I2 = (1/(2*a)) * sp.integrate(psi, (xi, x - a*t, x + a*t))
I3 = (1/(2*a)) * sp.integrate(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau))), (tau, 0, t))

u = I1 + I2 + I3

equation = f"""
\\begin{{align*}}
\\begin{{cases}}
    u_{{tt}}(x, t) &= {sp.latex(a**2)}^2u_{{xx}}(x, t) + {sp.latex(f)}, \\\\
    u(x, 0) &= {sp.latex(phi)}, \\\\
    u_t(x, 0) &= {sp.latex(psi)}.
\\end{{cases}}
\\end{{align*}}
"""
display(Math(equation))

display(Math(f"I_1 = \\frac{{1}}{{2}} \\left( \\left( {sp.latex(phi.subs(x, x + a * t))}\\right) + \\left( {sp.latex(phi.subs(x, x - a * t))}\\right) \\right)"))
display(Math(f"I_1 = \\frac{{1}}{{2}} \\left( {sp.latex(phi.subs(x, x + a * t) + phi.subs(x, x - a * t))} \\right)"))
display(Math(f"I_1 = \\frac{{1}}{{2}} \\left( {sp.latex(sp.simplify(phi.subs(x, x + a * t) + phi.subs(x, x - a * t)))} \\right)"))
display(Math(f"I_1 = {sp.latex((1/2) * (phi.subs(x, x + a * t) + phi.subs(x, x - a * t)))}"))
display(Math(f"I_1 = {sp.latex(sp.simplify((1/2) * (phi.subs(x, x + a * t) + phi.subs(x, x - a * t))))}"))

display(Math(f"I_2 = \\frac{{1}}{{2{sp.latex(a)}}} \\int_{{x - {sp.latex(a)} t}}^{{x + {sp.latex(a)} t}} {sp.latex(psi)} \\, d\\xi"))
display(Math(f"I_2 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.integrate(psi, (xi, x - a*t, x + a*t)))} \\Biggr|_{{x - {sp.latex(a)} t}}^{{x + {sp.latex(a)} t}} \\right)"))
display(Math(f"I_2 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.simplify(sp.integrate(psi, (xi, x - a*t, x + a*t))))} \\Biggr|_{{x - {sp.latex(a)} t}}^{{x + {sp.latex(a)} t}} \\right)"))
display(Math(f"I_2 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.integrate(psi, (xi, x - a*t, x + a*t)))} \\right)"))
display(Math(f"I_2 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.simplify(sp.integrate(psi, (xi, x - a*t, x + a*t))))} \\right)"))
display(Math(f"I_2 = {sp.latex((1/(2*a)) *sp.simplify(sp.integrate(psi, (xi, x - a*t, x + a*t))))}"))
display(Math(f"I_2 = {sp.latex(sp.simplify((1/(2*a)) *sp.simplify(sp.integrate(psi, (xi, x - a*t, x + a*t)))))}"))

display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\int_{{0}}^{{t}} \\int_{{x - {sp.latex(a)}(t - \\tau)}}^{{x + {sp.latex(a)}(t - \\tau)}} {sp.latex(f.subs({x: xi, t: tau}))} \\, d\\xi \\, d\\tau"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\int_{{0}}^{{t}} {sp.latex(sp.integrate(f.subs({x: xi, t: tau}), xi))} \\Biggr|_{{x - {sp.latex(a)}(t - \\tau)}}^{{x + {sp.latex(a)}(t - \\tau)}} \\, d\\tau"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\int_{{0}}^{{t}} {sp.latex(sp.simplify((sp.integrate(f.subs({x: xi, t: tau}), xi))))} \\Biggr|_{{x - {sp.latex(a)}(t - \\tau)}}^{{x + {sp.latex(a)}(t - \\tau)}} \\, d\\tau"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\int_{{0}}^{{t}} {sp.latex(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau))))} \\, d\\tau"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\int_{{0}}^{{t}} {sp.latex(sp.simplify(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau)))))} \\, d\\tau"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.integrate(sp.simplify(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau)))), tau))} \\Biggr|_{{0}}^{{t}} \\right)"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.simplify(sp.integrate(sp.simplify(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau)))), tau)))} \\Biggr|_{{0}}^{{t}} \\right)"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.integrate(sp.simplify(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau)))), (tau , 0 , t)))} \\right)"))
display(Math(f"I_3 = \\frac{{1}}{{2{sp.latex(a)}}} \\left( {sp.latex(sp.simplify(sp.integrate(sp.simplify(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau)))), (tau , 0 , t))))} \\right)"))
display(Math(f"I_3 = {sp.latex((1/(2*a)) * sp.integrate(sp.simplify(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau)))), (tau , 0 , t)))}"))
display(Math(f"I_3 = {sp.latex(sp.simplify((1/(2*a)) * sp.integrate(sp.simplify(sp.integrate(f.subs({x: xi, t: tau}), (xi, x - a*(t - tau), x + a*(t - tau)))), (tau , 0 , t))))}"))

display(Math(f"u(x,y) = {sp.latex(u)}"))
display(Math(f"u(x,y) = {sp.latex(sp.simplify(u))}"))
```
